## Numpy array and basics 
### Learning numpy


```python
import numpy as np
```

## Creating array list 


```python
arr_1d = np.array([1,2,3,4,5,6])
print("1D array : ",arr_1d)

arr_2d = np.array([[1,2,3,4],[1,2,3,4]])
print("2D array : ",arr_2d)
```

## List vs numpy array 

```python
py_list = [1,2,3,4]
print("Python list multiplication",py_list * 2)

np_array = np.array([1,2,3,4]) # element wise multiplication
print("Python array multiplication",np_array *2 )


import time 
start = time.time()
py_list = [i*2 for i in range(100000)]
print("\n operation time : ",time.time() - start)

start = time.time()
np_array = np.arange(100000)*2
print("\n operation time : ",time.time() - start)

```

## Creating array  from scratch


```python
zeros = np.zeros((3,4))
print("Zero array : ",zeros)

ones = np.ones((3,3))
print("ones array : ",ones)

full = np.full((2,2)  ,4)
print("Full : ",full)

random = np.random.random((2,3))
print("Random : ",random)

sequence = np.arange(0,12,2)
print("sequence : ",sequence)
```

## Vectors , matrix ,tensor

```python
vector = np.array([1,2,3,4])
print("Vector : ",vector)

matrix = np.array([[1,2,3,4],[1,3,4,5]])
print("Matrix : ",matrix)

tensor = np.array([[[1,2],[5,6]],
                    [[5,6],[7,8]]])
print("Tensor : ",tensor)
```

## Array properties 

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape : ",arr.shape)
print("dimesnion : ",arr.ndim)
print("Size : ",arr.size)
print("Datatype : ",arr.dtype)
```

## Array reshaping

```python
array = np.arange(12)
print("Original array : ",array)

reshaped_array = array.reshape((3,4))
print("Reshaped array : ",reshaped_array)

flatten = reshaped_array.flatten()
print("Flatten : ",flatten)

revaled = reshaped_array.ravel()
print("Raveled : ",revaled)

transpose = reshaped_array.T
print("Transpose : ",transpose)
```
