```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
```

```python
array1 = np.array([1,2,3,4,5])
array2 = np.random.rand(3,3)
array3 = np.zeros((4,4))
```

```python
np.save("array1.npy",array1)
np.save("array2.npy",array2)
np.save("array3.npy",array3)
```

```python
loaded_array1 = np.load("array1.npy")
print("Loaded array: ",loaded_array1)
```

```python
try:
    logo = mpimg.imread("numpy-logo.png")

    #Display the  logo
    plt.figure(figsize=(8,6))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)
    logo = logo / 255.0
    dark_logo = 1 -   logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Dark logo")
    plt.grid(False)

except FileNotFoundError:
    print("File not found.Logo missing")
```

```python

plt.figure(figsize=(7,7))
print(logo.min(),logo.max())
bright =  logo + 0.2
bright = np.clip(bright,0,1)
plt.subplot(121)
plt.imshow(bright)
plt.grid(False)
plt.title("Numpy logo")
plt.show()
```
