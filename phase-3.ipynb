```python
#Data Strutcure 
import numpy as np
import matplotlib.pyplot as plt 

sales_data = np.array(
    [[1,1150000,200000,240000,230000],
    [2,250000,200000,240000,200000],
    [3,350000,100000,240000,100000],
    [4,250000,230000,240000,400000],
    [5,750000,205000,240400,200000]]
)
```

```python
print("======= Zomato Sales data =======")
print("Sales data shapes " ,sales_data.shape)

print("Sample data for 1st 3 res t: ",sales_data[:3])

#Total sales per year
print("Total sales per year : ",np.sum(sales_data,axis=0))

yearly_total = np.sum(sales_data[:,1:],axis=0)
print("Yearly total : ",yearly_total)
print("Total : ",np.sum(yearly_total))

#Minimun and max sales per resturant
min_sales = np.min(sales_data[:,1:],axis=1)
print(min_sales)

#Max sales per year
max_sales = np.max(sales_data[:,1:],axis=0)
print("Max sales per year : ",max_sales)

#Avg sales per restaurant
avg_sales = np.mean(sales_data[:,1:],axis=1)
print("Avg sales per restaurant : ",avg_sales)

cumsum = np.cumsum(sales_data[:,1:],axis=1)
print("Cumulative sum : ",cumsum)

plt.figure(figsize=(6,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Avg cumulative sales per year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

vector_sum = vector1+vector2
print("\nVector sum : ",vector_sum)

print("multplication : ",vector1*vector2)

dot_product = np.dot(vector1,vector2)
print("Dot product : ",dot_product)

angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle btw 2 vector : ",np.degrees(angle))
```
